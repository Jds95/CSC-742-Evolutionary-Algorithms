```python
def sigmoid(Z):
    return 1 / (1 + np.exp(-Z))

def relu(Z):
    return np.maximum(0, Z)

def tanh(z):
    return np.tanh(z)

class NeuralNetwork:
    def __init__(self, bs_feature, bs_pos, bs_lemma, ws_feature, ws_pos, ws_lemma, act_fcn):
        """Create a neural network.

        Args:
            weights : a list of matrices that are L x L-1 in shape
            biases  : a list of matrices that are L x 1 in shape
            act_fcn : an activation function
        """
        self.bs_feature 
        self.bs_pos 
        self.bs_lemma 
        self.ws_feature
        self.ws_pos
        self.ws_lemma
        self.act_fcn = act_fcn

    def activate(self, X):
        """Feed the input X through the network.

        Args:
            X : a matrix that is `number of inputs` by `batch size` in shape
        """
        for b, w in zip(self.biases, self.weights):
            X = self.act_fcn(np.dot(w, X) + b)
        return X
```

```python
def fitness_fcn(lemma):
    inputs = np.array([lemma[:, cue]])
    outputs = np.array([lemma[:, feature, pos, lemma]])
    #     ann_outputs = [ann.activate(xor2_input) for xor2_input in xor2_inputs]
    #     ann_outputs = []
    #     for xor2_input in xor2_inputs:
    #         ann_outputs.append(ann.activate(xor2_input))

    ann_outputs = ann.activate(inputs)
    return np.mean(np.abs(ann_outputs - outputs))
```

```python
# A 1 input, 3 hidden node, 3 output network

num_cue = lemma.shape[0]

num_feature = lemma.shape[0]

num_pos = lemma.shape[0]

num_lemma = lemma.shape[0]

sizes = (num_cue, 3, num_feature, num_pos, num_lemma)

# Create a Lx1 matrix of bias values for each non-input layer
bs_feature = [np.random.randn(y, 1) for y in sizes[1:]]
print([b.shape for b in bs_feature])

bs_pos = [np.random.randn(y, 1) for y in sizes[1:]]
print([b.shape for b in bs_pos])

bs_lemma = [np.random.randn(y, 1) for y in sizes[1:]]
print([b.shape for b in bs_lemma])

# Create a LxL-1 matrix of weight values for each layer

ws_feature = [np.random.randn(y, 1) for y in sizes[1:]]
print([w.shape for w in ws_feature])

ws_pos = [np.random.randn(y, x) for x, y in zip(sizes[:-1], sizes[1:])]
print([w.shape for w in ws_pos])

ws_lemma = [np.random.randn(y, x) for x, y in zip(sizes[:-1], sizes[1:])]
print([w.shape for w in ws_lemma])

ann = NeuralNetwork(bs_feature, bs_pos, bs_lemma, ws_feature, ws_pos, ws_lemma, tanh)
fitness_fcn(ann)
```

```python
def create_ann_from_genome(genome, topology):

    # Variables for looping through the genome
    a = 0
    b = 0

    # Extract the bias values
    bs = []
    for layer_size in topology[1:]:
        b += layer_size
        bs.append(np.array(genome[a:b]).reshape((layer_size, 1)))
        a = b

    # Extract the weight values
    ws = []
    for layer_size, next_size in zip(topology[:-1], topology[1:]):
        b += layer_size * next_size
        ws.append(np.array(genome[a:b]).reshape((next_size, layer_size)))
        a = b

    return NeuralNetwork(ws, bs, genome[-1])


def create_ann_genome(topology, max_weight, act_fcns):

    num_biases = sum(topology[1:])
    num_weights = sum([x * y for x, y in zip(topology[:-1], topology[1:])])

    # Generate random biases and weights
    wm = max_weight
    genome = [
        np.random.uniform(-wm, wm) for _ in range(num_biases + num_weights)
    ]

    # Select an activation function at random
    genome += [np.random.choice(act_fcns)]

    return genome


class ANNIndividual:
    def __init__(self,
                 cue,
                 feature,
                 pos,
                 lemma,
                 topology,
                 fitness_fcn,
                 act_fcns=[sigmoid, tanh, relu],
                 max_weight=4):
        if genome == None:
            self.genome = create_ann_genome(topology, max_weight, act_fcns)
        else:
            self.genome = genome

        self.topology = topology

        ann = create_ann_from_genome(self.genome, self.topology)
        self.fitness = fitness_fcn(ann)

    def __repr__(self):
        s = f'Fitness={self.fitness:1.2f}'
        s += ' Genome=[' + ' '.join([f'{g:1.2f}' for g in self.genome[:-1]])
        s += f' {self.genome[-1]}]'
        return s
```
